```python
ir50web = pandas.read_csv("Tests/gms_bzorig_b3lyp_nosmd_opt-50.csv",header=1)
ir50web = ir50web.dropna(axis='columns', how='all')
ir50web
```

```python
ir10web = pandas.read_csv("Tests/gms_bzorig_b3lyp_nosmd_opt-10.csv",header=1)
ir10web = ir10web.dropna(axis='columns', how='all')
ir10web
```

```python
ir10or = pandas.read_csv("Tests/PlotCompare/Lorentzian/BzOrig/gms_bzorig_b3lyp_nosmd_ir_oth.csv",header=1)
ir10or = ir10or.dropna(axis='columns', how='all')
ir10or = ir10or.sort_values(by=['Freq'], ascending=True)
ir10or = ir10or.reset_index(drop=True)
ir10or.dtypes
```

```python
ir50web.plot()
ir10web.plot()
ir10or.plot()
```

```python
# Create a new figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each dataset
ax.plot(ir50web['Frequency (cm^-1)'], ir50web['Intensity (arbitrary units)'], label='ir50web')
ax.plot(ir10web['Frequency (cm^-1)'], ir10web['Intensity (arbitrary units)'], label='ir10web')
ax.plot(ir10or['Freq'], ir10or['Intensity'], label='ir10or')

# Customize the plot
ax.set_xlabel('Frequency')
ax.set_ylabel('Intensity')
ax.set_title('Comparison of ir50web, ir10web, and ir10or')
ax.legend()

# Display the plot
plt.show()
```
